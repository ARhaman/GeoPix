# 1. Facies Augmentation

```python
import os
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array, save_img, ImageDataGenerator

# Directories
original_dir = '/home/g202103050/Documents/A.Rahman/2-3D GAN part/2D/2D_New Eara/Real layers/orginal_facies__layer_images'
augmented_dir = '/home/g202103050/Documents/A.Rahman/2-3D GAN part/2D/2D_New Eara/Augumented layers/Facies_augmented'

os.makedirs(augmented_dir, exist_ok=True)

# Load images
def load_images(directory):
    images, filenames = [], []
    for filename in sorted(os.listdir(directory)):
        filepath = os.path.join(directory, filename)
        if os.path.isfile(filepath):
            img = load_img(filepath, color_mode='rgb')
            images.append(img_to_array(img))
            filenames.append(filename)
    return images, filenames

images, filenames = load_images(original_dir)

# ImageDataGenerator
datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Augment and save images
for i, (image, filename) in enumerate(zip(images, filenames)):
    base_name = os.path.splitext(filename)[0]
    image_batch = np.expand_dims(image, axis=0)
    aug_iter = datagen.flow(image_batch, batch_size=1, seed=i)
    for j in range(10):
        aug_img = next(aug_iter)[0].astype('uint8')
        save_path = os.path.join(augmented_dir, f"{base_name}_aug{j+1}.png")
        save_img(save_path, aug_img)
        print(f"Saved augmented facies image: {save_path}")
```
