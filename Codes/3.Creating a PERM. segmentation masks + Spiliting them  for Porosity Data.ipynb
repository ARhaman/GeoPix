```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Directories
input_dir = "/home/g202103050/Documents/A.Rahman/2-3D GAN part/2D/2D_New Eara/Augumented layers/Facies_augmented"
output_dir = "/home/g202103050/Documents/A.Rahman/2-3D GAN part/2D/2D_New Eara/Masks/Augmtd_facies__layer_images0000"
os.makedirs(output_dir, exist_ok=True)

def create_facies_mask(image):
    """
    Create semantic mask preserving geological patterns
    """
    img_array = np.array(image)
    mask = np.ones((*img_array.shape[:2], 3), dtype=np.uint8) * 0  # Initialize with black/dark background
    
    # Get grayscale intensity for pattern detection
    intensity = np.mean(img_array, axis=2)
    
    # Create masks based on color intensity patterns
    yellow_mask = np.logical_and(
        intensity > 200,  # Bright regions
        np.logical_and(img_array[:,:,0] > 240, img_array[:,:,1] > 240)
    )
    
    orange_mask = np.logical_and(
        intensity > 150,
        np.logical_and(img_array[:,:,0] > 200, img_array[:,:,1] > 140)
    )
    
    gray_mask = np.logical_and(
        intensity > 100,
        np.logical_and(
            np.abs(img_array[:,:,0] - img_array[:,:,1]) < 30,
            np.abs(img_array[:,:,1] - img_array[:,:,2]) < 30
        )
    )
    
    # Apply masks with distinct colors
    mask[yellow_mask] = [255, 255, 0]   # Bright yellow
    mask[orange_mask] = [255, 165, 0]   # Orange
    mask[gray_mask] = [255, 255, 255]   # White (changed from gray)
    
    # Set non-geological areas to dark blue background
    background = ~(yellow_mask | orange_mask | gray_mask)
    mask[background] = [0, 0, 128]  # Dark blue background
    
    return mask

def create_paired_image(original, mask):
    """
    Create paired image (original | mask) for pix2pix
    """
    # Create paired image (original | mask)
    paired = np.concatenate((original, mask), axis=1)  # Original on the left, mask on the right
    
    return paired

# Process images
for filename in sorted(os.listdir(input_dir)):
    if filename.endswith(('.png', '.jpg', '.jpeg')):
        # Load original image
        image_path = os.path.join(input_dir, filename)
        original_image = np.array(Image.open(image_path))

        # Create semantic mask
        semantic_mask = create_facies_mask(original_image)

        # Create paired image
        paired_image = create_paired_image(original_image, semantic_mask)

        # Save paired image with original size and high quality
        output_path = os.path.join(output_dir, f"paired_{filename}")
        Image.fromarray(paired_image).save(
            output_path, 
            format='PNG',  # Ensure it's saved as PNG
            quality=95,
            optimize=True
        )
        
        print(f"Processed: {filename}")

# Display examples
def display_examples(input_dir, output_dir, num_examples=3):
    plt.figure(figsize=(12, 4*num_examples))
    for i, filename in enumerate(sorted(os.listdir(output_dir))[:num_examples]):
        if filename.startswith('paired_'):
            paired_path = os.path.join(output_dir, filename)
            paired_img = plt.imread(paired_path)
            
            plt.subplot(num_examples, 1, i+1)
            plt.imshow(paired_img)
            plt.title(f"Paired Image {i+1} (Original | Mask)")
            plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Display some examples
display_examples(input_dir, output_dir)
```

# Spilit Dataset

```python
import os
import random
import shutil

# Paths
original_dir = '/home/g202103050/Documents/A.Rahman/2-3D GAN part/2D/2D_New Eara/Masks/Orginal_Sw_layer_images/'
augmented_dir = '//home/g202103050/Documents/A.Rahman/2-3D GAN part/2D/2D_New Eara/Masks/Aug_Sw_layer_images/'
output_dir = '/home/g202103050/Documents/A.Rahman/2-3D GAN part/2D/2D_New Eara/dataset_split/Sw/'

# Create output directories
train_dir = os.path.join(output_dir, 'train')
val_dir = os.path.join(output_dir, 'val')
test_dir = os.path.join(output_dir, 'test')

for directory in [train_dir, val_dir, test_dir]:
    os.makedirs(directory, exist_ok=True)

# Load all image file paths
original_images = [os.path.join(original_dir, f) for f in os.listdir(original_dir) if f.endswith('.png')]
augmented_images = [os.path.join(augmented_dir, f) for f in os.listdir(augmented_dir) if f.endswith('.png')]

# Combine original and augmented images
all_images = original_images + augmented_images
random.shuffle(all_images)

# Calculate split sizes
total_images = len(all_images)
train_count = int(0.7 * total_images)
val_count = int(0.15 * total_images)
test_count = total_images - train_count - val_count  # Ensure all are used

# Split datasets
train_images = all_images[:train_count]
val_images = all_images[train_count:train_count + val_count]
test_images = all_images[train_count + val_count:]

# Function to copy and rename images
def copy_and_rename_images(image_list, target_dir):
    for i, img_path in enumerate(image_list, start=1):  # Start counting from 1
        save_path = os.path.join(target_dir, f"{i}.jpg")
        shutil.copy(img_path, save_path)

# Copy images to respective folders with sequential naming
copy_and_rename_images(train_images, train_dir)
copy_and_rename_images(val_images, val_dir)
copy_and_rename_images(test_images, test_dir)

# Print results
print(f"Training set: {len(train_images)} images")
print(f"Validation set: {len(val_images)} images")
print(f"Testing set: {len(test_images)} images")
```
