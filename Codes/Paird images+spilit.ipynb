left facies vs right porosity

```python
import os
from PIL import Image

# Define directories
porosity_dir = '/home/g202103050/Documents/A.Rahman/2-3D GAN part/2D/2D_New Eara/Real layers/orginal_poro_layer_images'
facies_dir = '/home/g202103050/Documents/A.Rahman/2-3D GAN part/2D/2D_New Eara/Real layers/orginal_facies__layer_images'
output_dir = '/home/g202103050/Documents/A.Rahman/2-3D GAN part/2D/2D_New Eara/Facies to por or perm or Sw/org_fc vs por'

os.makedirs(output_dir, exist_ok=True)

# Debug: List all files in both directories
print("Available porosity images:")
print(os.listdir(porosity_dir))
print("\nAvailable facies images:")
print(os.listdir(facies_dir))

# Loop through each porosity image
for porosity_filename in os.listdir(porosity_dir):
    if porosity_filename.startswith('Poro_layer') and porosity_filename.endswith('.png'):
        porosity_path = os.path.join(porosity_dir, porosity_filename)

        # Extract the layer number from the porosity filename
        try:
            layer_number = porosity_filename.split('_')[2]  # Adjust based on actual structure
            print(f'\nExtracted layer number: {layer_number} from {porosity_filename}')

            # Construct the matching facies filename
            facies_filename = f'Facies_layer_{layer_number}_ultra_high_res.png'
            facies_path = os.path.join(facies_dir, facies_filename)
            print(f'Looking for facies file: {facies_filename}')

            # Check if the facies file exists
            if os.path.exists(facies_path):
                print(f'Matching facies file found: {facies_filename}')
                
                # Proceed with combining the images
                porosity_image = Image.open(porosity_path)
                facies_image = Image.open(facies_path)

                # Resize images to the same height
                new_height = min(porosity_image.height, facies_image.height)
                porosity_image = porosity_image.resize((porosity_image.width, new_height), Image.Resampling.LANCZOS)
                facies_image = facies_image.resize((facies_image.width, new_height), Image.Resampling.LANCZOS)

                # Create a new blank image with the combined width and same height
                combined_width = porosity_image.width + facies_image.width
                combined_image = Image.new('RGB', (combined_width, new_height))

                # Paste images side by side with facies on the left and porosity on the right
                combined_image.paste(facies_image, (0, 0))  # Facies on the left
                combined_image.paste(porosity_image, (facies_image.width, 0))  # Porosity on the right

                # Save the combined image
                combined_output_path = os.path.join(output_dir, f'combined_layer_{layer_number}.png')
                combined_image.save(combined_output_path, dpi=(300, 300), quality=95, format='PNG')

                print(f'Combined image saved: {combined_output_path}')
            else:
                print(f'No matching facies file found for layer {layer_number}')
        except IndexError as e:
            print(f'Error extracting layer number from {porosity_filename}: {e}')
```

```python
import os
import random
import shutil

# Paths
original_dir = '/home/g202103050/Documents/A.Rahman/2-3D GAN part/2D/2D_New Eara/Facies to por or perm or Sw/all_org_Sw vs Fc/'
augmented_dir = '/home/g202103050/Documents/A.Rahman/2-3D GAN part/2D/2D_New Eara/Facies to por or perm or Sw/all_Aug_Sw vs Fc/'
output_dir = '//home/g202103050/Documents/A.Rahman/2-3D GAN part/2D/2D_New Eara/Facies to por or perm or Sw/all_Sw vs Fc_dataset_split/'

# Create output directories
train_dir = os.path.join(output_dir, 'train')
val_dir = os.path.join(output_dir, 'val')
test_dir = os.path.join(output_dir, 'test')

for directory in [train_dir, val_dir, test_dir]:
    os.makedirs(directory, exist_ok=True)

# Load all image file paths
original_images = [os.path.join(original_dir, f) for f in os.listdir(original_dir) if f.endswith('.png')]
augmented_images = [os.path.join(augmented_dir, f) for f in os.listdir(augmented_dir) if f.endswith('.png')]

# Combine original and augmented images
all_images = original_images + augmented_images
random.shuffle(all_images)

# Calculate split sizes
total_images = len(all_images)
train_count = int(0.7 * total_images)
val_count = int(0.15 * total_images)
test_count = total_images - train_count - val_count  # Ensure all are used

# Split datasets
train_images = all_images[:train_count]
val_images = all_images[train_count:train_count + val_count]
test_images = all_images[train_count + val_count:]

# Function to copy and rename images
def copy_and_rename_images(image_list, target_dir):
    for i, img_path in enumerate(image_list, start=1):  # Start counting from 1
        save_path = os.path.join(target_dir, f"{i}.png")
        shutil.copy(img_path, save_path)

# Copy images to respective folders with sequential naming
copy_and_rename_images(train_images, train_dir)
copy_and_rename_images(val_images, val_dir)
copy_and_rename_images(test_images, test_dir)

# Print results
print(f"Training set: {len(train_images)} images")
print(f"Validation set: {len(val_images)} images")
print(f"Testing set: {len(test_images)} images")
```
